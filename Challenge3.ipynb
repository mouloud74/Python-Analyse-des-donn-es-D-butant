```python
import pandas as pd
```

```python
data = pd.read_csv('employee.csv')
```

+ creation d'une fonction qui retourn la moyenne(par défaut)
, d'une colonne ou plusieurs colonnes en fonction des autres colonnes
+ Modifiable

```python
def agg_colonnes(df,features,by_cols,methode='mean'):
    return df[features].groupby(by_cols).agg(methode)
```

+ # Appliquons notre fonction :

```python
features = ['Age','Department','EducationField']
by_cols = 'Department'
agg_colonnes(data,features,by_cols)
```

```python
# Changeons le paramètre methode='count'
agg_colonnes(data,['EducationField','Department'],by_cols,methode='count')
```

```python
# Age moyenne par fonction et par genre
agg_colonnes(data,['Age','Gender','EducationField'],['Gender', 'EducationField'],methode='mean')
```
