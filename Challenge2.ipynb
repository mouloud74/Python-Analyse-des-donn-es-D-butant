+ Grouper le dataset par département et calculer la moyenne

```python
data.groupby('Department').mean()# ou agg('mean')
```

```python
# On prend juste les deux variables :age et department
cols = ['Age','Department']
data[cols].groupby('Department').agg('mean')
```

```python
# On rajoute la colonne des revenus mensuel : MonthlyIncome
cols = ['Age','MonthlyIncome','Department']
data[cols].groupby('Department').agg('mean')
```

```python
# On groupe par department et par genre
# On ajoute un style à la sortie.
cols = ['Age','MonthlyIncome','Department','Gender']
data[cols].groupby(['Department','Gender']).agg('mean').style.bar(color=['gold'])
```
