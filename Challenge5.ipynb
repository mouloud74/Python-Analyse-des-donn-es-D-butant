```python
# Liste des colonnes qui ont au moins une valeur manquante.
cols_missing_values = []
for col in data.columns:
    if data[col].isnull().any():
        cols_missing_values.append(col)
print(f'Voici la liste des colonnes qui présentent des valeurs manquantes :{cols_missing_values}')
```

```python
data.columns
```

```python
cols = ['model_year', 'origin']
data[cols].tail()
```

```python
# La moyenne mpg des modeles 70 à 75 enfonction des pays.
df = data.groupby(by=cols).mean()['mpg'].loc[70:75].unstack()
df
```

```python
# Arrondir les colonnes à 2 chiffres après la virgule
df = df.round(2)
df
```

# Utilisation de pivot_table

```python
pivot_moyenne_mpg = pd.pivot_table(data=moyenne_mpg,index='model_year',values="mpg",columns="origin",
               margins=True,margins_name="Total")
```

```python
# Afficher les cinq premières lignes du dataset
pivot_moyenne_mpg.iloc[0:5]
```

```python
# Réinitialiser l'index
pivot_moyenne_mpg.reset_index(inplace=True)
```

```python
pivot_moyenne_mpg.head() # .iloc[0:5]
```

```python
# melt permet de passer du format large au format long du dataframe
pivot_melt = pivot_moyenne_mpg.melt(id_vars= "model_year", value_vars=["europe","japan","usa"], var_name = "origine", value_name = "mpg")
pivot_melt.loc[0:13] # Les 13 premières lignes
```
